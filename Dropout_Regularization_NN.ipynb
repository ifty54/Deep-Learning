#Dropout Regularization Technique In Deep Neural Network

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
%matplotlib inline
import seaborn as sns
```

```python
path = '/content/sonar_dataset.csv'
df = pd.read_csv(path, header= None)
```

```python
df.head()
```

```python
df.shape
```

```python
#Checking NaN Values
df.isna().sum()
```

```python
df.columns
```

```python
features = df.drop(60, axis = 'columns')
target = df[60]
```

```python
target
```

```python
target = pd.get_dummies(target, drop_first = True)
target.sample(5)
```

M becomes 0 and R = 1

```python
target.value_counts()
```

M = 111, R = 97

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.25, random_state=1)
```

```python
X_train.head()
```

```python
model = keras.Sequential([
    keras.layers.Dense(60, input_dim=60, activation='relu'),
    keras.layers.Dense(30, activation='relu'),
    keras.layers.Dense(15, activation='relu'),
    keras.layers.Dense(1, activation='sigmoid')
])

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

model.fit(X_train, y_train, epochs=100, batch_size=8)
```

```python
model.evaluate(X_test, y_test)
```

```python
y_pred = model.predict(X_test).reshape(-1)
print(y_pred[:10])

# round the values to nearest integer ie 0 or 1
y_pred = np.round(y_pred)
print(y_pred[:10])
```

```python
from sklearn.metrics import confusion_matrix , classification_report

print(classification_report(y_test, y_pred))
```

##Now, Applying Dropout method instead of doing flat-wise calculation in the Neural Networks

```python
model_dropout = keras.Sequential([
    keras.layers.Dense(60, input_dim=60, activation='relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(30, activation='relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(15, activation='relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(1, activation='sigmoid')
])

model_dropout.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

model_dropout.fit(X_train, y_train, epochs=100, batch_size=8)
```

```python
model_dropout.evaluate(X_test, y_test)
```

```python
y_pred = model_dropout.predict(X_test).reshape(-1)
print(y_pred[:10])

# round the values to nearest integer ie 0 or 1
y_pred = np.round(y_pred)
print(y_pred[:10])
```

```python
from sklearn.metrics import confusion_matrix , classification_report

print(classification_report(y_test, y_pred))
```

## **Result**: test accuracy increased from 0.77 to 0.81 by using Dropout Regularization. However, dropout doesn't always suit the model.
